```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Load the CSV file
data = pd.read_csv("ifood_df.csv")

# Preview the data
data.head()
```

```python
data.drop(columns=['ID', 'Z_CostContact', 'Z_Revenue'], inplace=True, errors='ignore')
```

```python
data.columns
```

```python
# Use the already existing feature
data['Customer_Days'].head()
```

```python
spend_columns = ['MntWines', 'MntFruits', 'MntMeatProducts',
                 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

data['TotalSpend'] = data[spend_columns].sum(axis=1)
```

```python
selected_features = [
    'Income', 'Customer_Days', 'Recency', 'MntTotal', 'MntRegularProds',
    'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases'
]
```

```python
X = data[selected_features].copy()
```

```python
X = X.dropna()  # or use fillna if appropriate
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Elbow Method to find optimal K
inertia = []
k_range = range(1, 11)  # Check for clusters from 1 to 10

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)  # Fit the model
    inertia.append(kmeans.inertia_)  # Save inertia

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
# Fit the KMeans model with the chosen number of clusters (e.g., k=4)
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Check the cluster centers
print("Cluster centers:\n", kmeans.cluster_centers_)
```

```python
# Visualize the distribution of TotalSpend by Cluster
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='MntTotal', data=data)
plt.title('Total Spend by Cluster')
plt.show()

# Alternatively, create scatter plots to show cluster separation (e.g., Total Spend vs Recency)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MntTotal', y='Recency', hue='Cluster', data=data, palette='viridis', s=100)
plt.title('Customer Segments (Clusters) by Total Spend and Recency')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Evaluate clustering with Silhouette Score
sil_score = silhouette_score(X_scaled, data['Cluster'])
print(f'Silhouette Score: {sil_score}')
```

```python
# Select numeric features to compare
features_to_plot = [
    'Income', 'Customer_Days', 'Recency', 'MntTotal',
    'MntRegularProds', 'NumWebPurchases', 'NumStorePurchases'
]

# Group by cluster and compute mean
cluster_summary = data.groupby('Cluster')[features_to_plot].mean().reset_index()

# Plot
plt.figure(figsize=(14, 8))
cluster_summary.set_index('Cluster').T.plot(kind='bar', figsize=(12, 6))
plt.title("Average Feature Values by Cluster")
plt.ylabel("Mean Value")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title="Cluster")
```
